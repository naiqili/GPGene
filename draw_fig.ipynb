```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# 50 gene

```python
nnet=5
nn=10

genie3pr, gpgeneLpr, gpgeneP2pr = np.zeros((nnet, nn)), np.zeros((nnet, nn)), np.zeros((nnet, nn))
genie3roc, gpgeneLroc, gpgeneP2roc = np.zeros((nnet, nn)), np.zeros((nnet, nn)), np.zeros((nnet, nn))

with open('/home/linaiqi/Lab/data/gene/tmp2/result_genie3_50gene.txt') as fin:
    for l in fin:
        v = l.strip().split()
        #print(v)
        genie3pr[int(v[2])-1, int(v[4])//100-1] = float(v[8])
        genie3roc[int(v[2])-1, int(v[4])//100-1] = float(v[11])

genie3pr_mu, genie3pr_std = np.mean(genie3pr, axis=0), np.std(genie3pr, axis=0)
genie3roc_mu, genie3roc_std = np.mean(genie3roc, axis=0), np.std(genie3roc, axis=0)

with open('/home/linaiqi/Lab/data/gene/tmp2/result_gpgeneL_50gene.txt') as fin:
    for l in fin:
        v = l.strip().split()
        #print(v)
        gpgeneLpr[int(v[2])-1, int(v[4])//100-1] = float(v[8])
        gpgeneLroc[int(v[2])-1, int(v[4])//100-1] = float(v[11])

gpgeneLpr_mu, gpgeneLpr_std = np.mean(gpgeneLpr, axis=0), np.std(gpgeneLpr, axis=0)
gpgeneLroc_mu, gpgeneLroc_std = np.mean(gpgeneLroc, axis=0), np.std(gpgeneLroc, axis=0)

with open('/home/linaiqi/Lab/data/gene/tmp/result_gpgeneP2_50gene.txt') as fin:
    for l in fin:
        v = l.strip().split()
        #print(v)
        gpgeneP2pr[int(v[2])-1, int(v[4])//100-1] = float(v[8])
        gpgeneP2roc[int(v[2])-1, int(v[4])//100-1] = float(v[11])

gpgeneP2pr_mu, gpgeneP2pr_std = np.mean(gpgeneP2pr, axis=0), np.std(gpgeneP2pr, axis=0)
gpgeneP2roc_mu, gpgeneP2roc_std = np.mean(gpgeneP2roc, axis=0), np.std(gpgeneP2roc, axis=0)
```

```python
xs=[int(100*(k+1)) for k in range(10)]

plt.figure()
plt.errorbar([x+10 for x in xs], gpgeneLpr_mu, gpgeneLpr_std, fmt='-', label='GPGene-linear', capsize=5)
plt.errorbar([x+0 for x in xs], gpgeneP2pr_mu, gpgeneP2pr_std, fmt=':', label='GPGene-poly2', capsize=5)
plt.errorbar([x-10 for x in xs], genie3pr_mu, genie3pr_std, fmt='--', label='GENIE3', capsize=5)
plt.legend(); plt.grid(); plt.xlabel('# of training instances'); plt.ylabel('AUPR'); 
```

```python
plt.figure()
plt.errorbar([x+10 for x in xs], gpgeneLroc_mu, gpgeneLroc_std, fmt='-', label='GPGene-linear', capsize=5)
plt.errorbar([x+0 for x in xs], gpgeneP2roc_mu, gpgeneP2roc_std, fmt=':', label='GPGene-poly2', capsize=5)
plt.errorbar([x-10 for x in xs], genie3roc_mu, genie3roc_std, fmt='--', label='GENIE3', capsize=5)
plt.legend(); plt.grid(); plt.xlabel('# of training instances'); plt.ylabel('AUROC'); 
```
